# Differential Expression Analysis

## Overview

The purpose of this notebook is to investigate the differential expression results of cell-line derived HIV-infected data, and the effect of the batch effect removal algorithm on these genes.

## Computation

```python
# Get infiles
infiles = ['f5-geneset_enrichment.dir/podocyte_cell_line-vst_corrected-differential_expression_geneset_enrichment.txt', 'f5-geneset_enrichment.dir/podocyte_cell_line-vst_corrected-differential_expression_geneset_ids.txt'] 

# Read data
enrichmentDataframe = pd.read_table(infiles[0])
genelistDataframe = pd.read_table(infiles[1])

# Fix timepoints
enrichmentDataframe['timepoint'] = [x.replace('h','')+'h' for x in enrichmentDataframe['timepoint']]
```

### Code

```python
# Enrichment data
enrichmentData = {}

# Add data
for direction in ['upregulated', 'downregulated']:
    enrichmentDataframeSubset = enrichmentDataframe.loc[enrichmentDataframe['direction'] == direction, :]
    enrichmentDataframeCast = enrichmentDataframeSubset.pivot(index='term_name', columns='timepoint', values='combined_score').fillna(0).loc[:,['6h','12h','24h','48h']]
    sortedTerms = enrichmentDataframeCast.apply(sum, 1).sort_values(ascending=False).index
    enrichmentData[direction] = enrichmentDataframeCast.loc[sortedTerms, :]
    
upregulatedEnrichmentData = enrichmentData['upregulated']
downregulatedEnrichmentData = enrichmentData['downregulated']
```

## Plots

```python
%%R -i upregulatedEnrichmentData,downregulatedEnrichmentData
plot_enrichment <- function(enrichmentDataframe, direction, termName, main=NULL) {
    
    # Fix colnames
    colnames(enrichmentDataframe) <- gsub('X', '', colnames(enrichmentDataframe))
    
    # Get plot dataframe
    barplot(rev(as.numeric(enrichmentDataframe[termName,])),
            names.arg=rev(colnames(enrichmentDataframe)),
            main = main,
            xlab = 'Enrichment Score',
            ylab = 'Timepoint',
            las = 2,
            horiz=TRUE)
}
```

```python
%%R -h 500 -w 800
par(mfrow=c(2,2))
plot_enrichment(upregulatedEnrichmentData, 'upregulated', termName='response to virus (GO:0009615)')
plot_enrichment(upregulatedEnrichmentData, 'upregulated', termName='type I interferon signaling pathway (GO:0060337)')
plot_enrichment(downregulatedEnrichmentData, 'downregulated', termName='negative regulation of protein phosphorylation (GO:0001933)')
plot_enrichment(downregulatedEnrichmentData, 'downregulated', termName='DNA replication-independent nucleosome assembly (GO:0006336)')
```
